```python
import pandas
import matplotlib.pyplot as plt

df25 = pandas.read_csv('25.csv')
df25 = df25.sort_values(by="moved_tiles", ascending=True).reset_index(drop=True)
df25
```

```python
# Második ábra: steps és steps_tile
plt.figure(figsize=(10, 6))
plt.plot(df25.index, df25["steps"], label="Steps", marker="o")
plt.plot(df25.index, df25["steps_tile"], label="Steps per Tile", marker="s")
plt.xlabel("Time")
plt.ylabel("Values")
plt.title("Steps and Steps per Tile (25)")
plt.legend()
plt.grid()
plt.show()
```

```python
# Harmadik ábra: moved_tiles
plt.figure(figsize=(10, 6))
plt.plot(df25.index, df25["moved_tiles"], label="Moved Tiles (25)", color="orange")
plt.xlabel("Time")
plt.ylabel("Values")
plt.title("Moved Tiles (25)")
plt.legend()
plt.grid()
plt.show()
```

```python
# Eloszlásdiagram az "steps" oszlophoz
plt.figure(figsize=(10, 6))
plt.hist(df25["steps"], bins=20, color="blue", alpha=0.7, label="Steps")
plt.xlabel("Steps")
plt.ylabel("Frequency")
plt.title("Distribution of Steps (25)")
plt.legend()
plt.grid()
plt.show()

# Eloszlásdiagram a "steps_tile" oszlophoz
plt.figure(figsize=(10, 6))
plt.hist(df25["steps_tile"], bins=20, color="green", alpha=0.7, label="Steps per Tile")
plt.xlabel("Steps per Tile")
plt.ylabel("Frequency")
plt.title("Distribution of Steps per Tile (25)")
plt.legend()
plt.grid()
plt.show()

# Eloszlásdiagram a "moved_tiles" oszlophoz
plt.figure(figsize=(10, 6))
plt.hist(df25["moved_tiles"], bins=20, color="orange", alpha=0.7, label="Moved Tiles")
plt.xlabel("Moved Tiles")
plt.ylabel("Frequency")
plt.title("Distribution of Moved Tiles (25)")
plt.legend()
plt.grid()
plt.show()
```

```python
# Eloszlásdiagramok egymás mellett (Steps, Steps per Tile, Moved Tiles)
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Steps oszlop eloszlása
axes[0].hist(df25["steps"], bins=20, color="blue", alpha=0.7, label="Steps")
axes[0].set_xlabel("Steps")
axes[0].set_ylabel("Frequency")
axes[0].set_title("Distribution of Steps (25)")
axes[0].legend()
axes[0].grid()

# Steps per Tile oszlop eloszlása
axes[1].hist(df25["steps_tile"], bins=20, color="green", alpha=0.7, label="Steps per Tile")
axes[1].set_xlabel("Steps per Tile")
axes[1].set_title("Distribution of Steps per Tile (25)")
axes[1].legend()
axes[1].grid()

# Moved Tiles oszlop eloszlása
axes[2].hist(df25["moved_tiles"], bins=20, color="orange", alpha=0.7, label="Moved Tiles")
axes[2].set_xlabel("Moved Tiles")
axes[2].set_title("Distribution of Moved Tiles (25)")
axes[2].legend()
axes[2].grid()

# Térközök optimalizálása a szebb megjelenés érdekében
plt.tight_layout()
plt.show()
```
